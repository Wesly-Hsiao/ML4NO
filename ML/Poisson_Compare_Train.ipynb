```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_IO = np.load('/work/NuFit_IO.npz')
data_NO = np.load('/work/NuFit_NO.npz')
data = np.load('/work/n1000000_0804_all_flat.npz')

data_all = np.column_stack([data['ve_dune'], data['vu_dune'], data['vebar_dune'], data['vubar_dune'], data['ve_t2hk'], data['vu_t2hk'], data['vebar_t2hk'], data['vubar_t2hk']])
data_IO_all = np.column_stack([data_IO['ve_dune'], data_IO['vu_dune'], data_IO['vebar_dune'], data_IO['vubar_dune'], data_IO['ve_t2hk'], data_IO['vu_t2hk'], data_IO['vebar_t2hk'], data_IO['vubar_t2hk']])[0]
data_NO_all = np.column_stack([data_NO['ve_dune'], data_NO['vu_dune'], data_NO['vebar_dune'], data_NO['vubar_dune'], data_NO['ve_t2hk'], data_NO['vu_t2hk'], data_NO['vebar_t2hk'], data_NO['vubar_t2hk']])[0]
```

```python
generate = 1000
num_of_bins = len(data_IO_all)
data_IO_create_poisson = np.random.poisson(data_IO_all, size = (generate, num_of_bins))
data_NO_create_poisson = np.random.poisson(data_NO_all, size = (generate, num_of_bins))
```

```python
data_max = data_all.max(axis = 0)
data_min = data_all.min(axis = 0)

plt.figure(figsize=(15, 9))
for i in range(generate):
    plt.plot(data_IO_create_poisson[i, ], 'b.', alpha = 0.1)
    plt.plot(data_NO_create_poisson[i, ], 'g.', alpha = 0.1)
plt.plot(data_max, label = 'max')
plt.plot(data_min, label = 'min')
plt.xlabel('bin')
plt.ylabel('number')
plt.ylim(0, 1000)
plt.legend()
plt.title('data range')

plt.show()
```

```python
data_IO_out = np.count_nonzero(np.logical_or(np.greater(data_IO_create_poisson, data_max), np.less(data_IO_create_poisson, data_min)), axis = 0)
data_NO_out = np.count_nonzero(np.logical_or(np.greater(data_NO_create_poisson, data_max), np.less(data_NO_create_poisson, data_min)), axis = 0)
```

```python
data_IO_out = data_IO_out/generate
data_NO_out = data_NO_out/generate
```

```python
plt.plot(data_IO_out)
plt.plot(data_NO_out)
plt.xlabel('bin')
plt.ylabel('ratio of poisson data out of range')
plt.title('{} times of poisson'.format(generate))
plt.show()
```
