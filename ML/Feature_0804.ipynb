```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.load('/work/n1000000_0804_all_flat.npz')
features = data.files
```

```python
features = ['ve_dune', 'vu_dune', 'vebar_dune', 'vubar_dune', 've_t2hk', 'vu_t2hk', 'vebar_t2hk', 
            'vubar_t2hk', 'theta12', 'theta13', 'theta23', 'delta', 'sdm', 'ldm', 'octant', 'cpv', 'mo']
classes = ['theta12', 'theta13', 'theta23', 'delta', 'sdm', 'ldm']
data_all = np.column_stack([data['ve_dune'], data['vu_dune'], data['vebar_dune'], data['vubar_dune'], data['ve_t2hk'], data['vu_t2hk'], data['vebar_t2hk'], data['vubar_t2hk']])
```

```python
for c in classes:
    plt.hist(data[c], bins = 100)
    plt.title(c)
    plt.show()
```

```python
data_max = data_all.max(axis = 0)
data_min = data_all.min(axis = 0)
```

```python
plt.figure(figsize=(15, 9))
plt.plot(data_max, label = 'max')
plt.plot(data_min, label = 'min')
plt.xlabel('bin')
plt.ylabel('number')
plt.ylim(0, 100)
plt.legend()
plt.title('data range')
plt.show()
```

```python
corr_data = pd.DataFrame(data['ve_dune'])
corr_M = corr_data.corr()
```

```python
plt.imshow(corr_M)
plt.colorbar()
plt.clim(0, 1)
plt.title('ve_dune 50bins correlation matrix')
```
