```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
#總資料組數
N = 121
```

```python
delta_cp = []
OFF_DU_40 = []
OFF_DU_50 = []
OFF_T2_40 = []
OFF_T2_50 = []
OFF_DUT2_40 = []
OFF_DUT2_50 = []
ON_DU_40 = []
ON_DU_50 = []
ON_T2_40 = []
ON_T2_50 = []
ON_DUT2_40 = []
ON_DUT2_50 = []
```

```python
#Import txt檔
f = open("sample_chi.txt")
```

```python
for i in range(N):
    s = f.readline().split()
    array = []
    for j in range(len(s)) :
        array.append(float(s[j])) 
    delta_cp.append(array[0])
    OFF_DU_40.append(array[1])    
    OFF_DU_50.append(array[2])
    OFF_T2_40.append(array[3])   
    OFF_T2_50.append(array[4])
    OFF_DUT2_40.append(array[5])   
    OFF_DUT2_50.append(array[6])
    ON_DU_40.append(array[7]) 
    ON_DU_50.append(array[8]) 
    ON_T2_40.append(array[9]) 
    ON_T2_50.append(array[10]) 
    ON_DUT2_40.append(array[11]) 
    ON_DUT2_50.append(array[12]) 
```

```python
delta_cp = np.array(delta_cp)
OFF_DU_40 = np.array(OFF_DU_40)
OFF_DU_50 = np.array(OFF_DU_50)
OFF_T2_40 = np.array(OFF_T2_40)
OFF_T2_50 = np.array(OFF_T2_50)
OFF_DUT2_40 = np.array(OFF_DUT2_40)
OFF_DUT2_50 = np.array(OFF_DUT2_50)
ON_DU_40 = np.array(ON_DU_40)
ON_DU_50 = np.array(ON_DU_50)
ON_T2_40 = np.array(ON_T2_40)
ON_T2_50 = np.array(ON_T2_50)
ON_DUT2_40 = np.array(ON_DUT2_40)
ON_DUT2_50 = np.array(ON_DUT2_50)
```

```python
plt.figure(figsize = (10,7))
plt.xlabel(r'True $\delta / \pi $', fontsize=15)
plt.ylabel(r'$\sqrt{\Delta\chi^2} $', fontsize=15)
plt.ylim(0, 16)
y0 = ON_DU_40**0.5
y1 = ON_DU_50**0.5
y2 = ON_T2_40**0.5
y3 = ON_T2_50**0.5
y4 = ON_DUT2_40**0.5
y5 = ON_DUT2_50**0.5
x = delta_cp/180
plt.plot(x,y0,'g',x,y1,'g',x,y2,'b',x,y3,'b',x,y4,'r',x,y5,'r')
plt.fill_between(x, y0, y1, where=y0 >= y1, facecolor='g', alpha = 0.3,label = 'DUNE')
plt.fill_between(x, y2, y3, where=y2 >= y3, facecolor='b', alpha = 0.3,label = 'T2HK')
plt.fill_between(x, y4, y5, where=y4 >= y5, facecolor='r', alpha = 0.3,label = 'DUNE+T2HK')
plt.legend()
plt.title('sys_ON', fontsize=15)
```
