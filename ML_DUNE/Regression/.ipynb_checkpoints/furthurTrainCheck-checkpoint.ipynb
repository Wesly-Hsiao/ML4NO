```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('/home/ML4NO/ML_DUNE/Regression/models_furthurTrain/0910_theta23_3_1_result.npy', allow_pickle=True)
data = data.item()
```
