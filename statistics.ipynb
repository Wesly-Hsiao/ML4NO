```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy
import os
```

```python
datas = []
for root, dirs, files in os.walk('/home/ML4NO/Data/data0902npz'):
    datas.append(files)
```

```python
data = np.load('n4963_q_cp_angle10.npz')
a = data['q_cpc'][::2]
b = data['q_cpv']
lines = np.load('cpv_sensitivity_globes.npz')
labels = lines.files[1:4]
```

```python
chi_square_list = []
for angle in range(90):
    data = np.load('/home/ML4NO/Data/data0902npz/n10000_q_cp_angle{}.npz'.format(angle))
    a = data['q_cpc'][::2]
    b = data['q_cpv']

    bins = np.linspace(min(a), max(b), 100)
    numa = len(a)
    numb = len(b)
    abin = plt.hist(a, bins=bins, alpha=0.7, label='a')
    bbin = plt.hist(b, bins=bins, alpha=0.7, label='b')
    cut = 0.50

    suma = 0
    for i, aval in enumerate(abin[0]):
        suma += aval
        if suma/numa >= 1-cut:
            acut = i
            break
    # plt.axvline(abin[1][acut], c='k')

    sumb = 0
    for i, bval in enumerate(bbin[0]):
        sumb += bval
        if sumb/numb >= cut:
            bcut = i
            break
    plt.axvline(bbin[1][bcut], c='k', label = 'beta = 0.5')

    plt.xlabel('T')
    plt.legend()
    plt.title('n10000_q_cp_angle{}'.format(angle))
    plt.show()
    pvalue = sum(abin[0][bcut:])/sum(abin[0])
    print('pvalue =', pvalue)
    chi = scipy.special.erfinv((1-2*pvalue))*np.sqrt(2)
    print('chi =', chi)
    chi_square = chi**2
    print('chi_square =', chi_square)
    chi_square_list.append(chi_square)
```

```python
plt.figure(figsize=(8,8))
for label in labels:
    plt.plot(lines['delta_cp'], lines[label], label = label)
plt.scatter([i for i in range(90)], chi_square_list, c='k')
plt.xlim(0,20)
plt.ylim(0,15)
plt.xlabel('True deltacp')
plt.ylabel('delta chi-square')
plt.legend()
plt.show()
```

```python
rsuma = []
rsumb = []
r = np.linspace(0, 1, 1000)
for rcut in r:
    suma = 0
    for i, aval in enumerate(abin[0]):
        suma += aval
        if suma/numa >= 1-rcut:
            rsuma.append(abin[1][i])
            break

    sumb = 0
    for i, bval in enumerate(bbin[0]):
        sumb += bval
        if sumb/numb >= rcut:
            rsumb.append(bbin[1][i])
            break
```

```python
plt.plot(rsuma, r)
plt.plot(rsumb, r)
plt.show()
```
